```python
plt.figure(figsize=(14, 7))
sns.barplot(x='Total Sales', y='Phone Models', data=df.sort_values('Total Sales', ascending=False), palette='mako')
plt.title('Общи продажби на различни модели телефони през 2024 година')
plt.xlabel('Общи продажби')
plt.ylabel('Модел телефони')
plt.show()
```

```python
df['Brand'] = df['Phone Models'].str.split().str[0]

brand_sales = df.groupby('Brand').sum().iloc[:, 1:].sum(axis=1)

top_5_brands = brand_sales.nlargest(5)

print(top_5_brands)

plt.figure(figsize=(8, 4))
plt.pie(top_5_brands, labels=top_5_brands.index, autopct='%1.1f%%', startangle=140)
plt.title('The 5 best brands mobile phones sells in 2024')
plt.axis('equal')
plt.show()
```

```python
df['Brand'] = df['Phone Models'].str.split().str[0]

brand_sales = df.groupby('Brand').sum().iloc[:, 1:].sum(axis=1)

low_5_brands = brand_sales.nsmallest(5)

print(low_5_brands)

plt.figure(figsize=(8, 4))
plt.pie(low_5_brands, labels=low_5_brands.index, autopct='%1.1f%%', startangle=140)
plt.title('The 5 of least sells brand phones in 2024')
plt.axis('equal')
plt.show()
```

```python
df = pd.DataFrame(data)

for index, row in df.iterrows():
    plt.figure(figsize=(10, 6))
    
    sales_values = row[1:].astype(float).values
    months = df.columns[1:]

    sns.barplot(x=months, y=sales_values, palette='viridis')
    plt.title(row['Phone Models'])
    plt.xlabel('Month')
    plt.ylabel('Sells')
    
   
    for i, v in enumerate(sales_values):
        plt.text(i, v + 5, str(int(v)), ha='center', va='bottom', fontsize=8)  # Корекция на позицията и стойността на текста
    
    plt.xticks(rotation=45, fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
df = pd.DataFrame(data)
df_melted = pd.melt(df, id_vars=['Phone Models'], var_name='Month', value_name='Sales')

print(df_melted.dtypes)

df_melted['Sales'] = pd.to_numeric(df_melted['Sales'], errors='coerce')

max_sales_per_model = df_melted.groupby('Phone Models')['Sales'].max()
min_sales_per_model = df_melted.groupby('Phone Models')['Sales'].min()

heatmap_data = pd.DataFrame({
    'Max Sales': max_sales_per_model,
    'Max Month': df_melted.loc[df_melted.groupby('Phone Models')['Sales'].idxmax(), 'Month'].values,
    'Min Sales': min_sales_per_model,
    'Min Month': df_melted.loc[df_melted.groupby('Phone Models')['Sales'].idxmin(), 'Month'].values
}).reset_index()

heatmap_pivot = heatmap_data.pivot_table(index='Phone Models', columns='Max Month', values='Max Sales')

plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_pivot, annot=True, cmap='YlGnBu', cbar=True, fmt=".0f")
plt.title('The strongest month sells for every brand phone')
plt.xlabel('Month')
plt.ylabel('Brand Phone')
plt.show()
```

```python
df = pd.DataFrame(data)

df_melted = pd.melt(df, id_vars='Phone Models', var_name='Month', value_name='Sales')

plt.figure(figsize=(18, 12))

sns.barplot(x='Month', y='Sales', hue='Phone Models', data=df_melted, dodge=False)

plt.title('Sells of mobile phones by months and brands (January - December 2024)')
plt.xlabel('Month')
plt.ylabel('Sells')

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
monthly_totals = df.drop(columns=['Phone Models']).sum().reset_index()
monthly_totals.columns = ['Month', 'Total Sales']

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_totals, x='Month', y='Total Sales', marker='o')

plt.xticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Combine Sells')
plt.title('Combine sells of phones by year 2024 г.')

plt.tight_layout()
plt.show()
```
